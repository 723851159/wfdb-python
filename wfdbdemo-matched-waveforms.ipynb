# Matched waveform example

This notebook loads and plots waveforms from the MIMIC matched waveform database. These waveforms have been matched to clinical data in the MIMIC Clinical database.

```python
import numpy as np
from wfdb import readsignal
from wfdb import readannot
from wfdb import plotwfdb
```

```python
# Demo 1 - Read the wfdb file, plot the last part of the signals, and print the metadata

sig, fields=readsignal.rdsamp('sampledata/matched/s25047/s25047-2704-05-04-10-44')
fields
```

```python
plotwfdb.plotsigs(sig[80000:160000,:], fields, title='s25047/s25047-2704-05-04-10-44') 
```
